I need to merge the two dataframes i.e. to paste the columns of the previous dataframe onto my energe_df.

```python
# Helper function to match index and assign value
def add_col(energy_df, weather_df, col):
    # Create a new column with default empty strings
    energy_df[col] = ''
    
    # Get the list of indexes
    energy_indexes = energy_df.index.tolist()
    weather_indexes = weather_df.index.tolist()
    
    
    # Loop over the index list
    for row in energy_indexes:
        if row in weather_indexes:
            energy_df.at[row, col] = weather_df.loc[row, col]

    return energy_df

```

```python
energy_df = add_col(energy_df, weather_df,'Acorn')
```
